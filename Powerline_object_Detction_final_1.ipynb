```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.editor import VideoFileClip

# Specify the path to your long video
video_path = "/content/runs/detect/predict2/hardhat.avi"

# Get the duration of the original video
original_clip = VideoFileClip(video_path)
video_duration = original_clip.duration

# Define the start times in seconds for each segment
start_times = [0, 120, 240]

# Initialize a list to store the video segments
clips = []

# Split the video into smaller segments and save them as separate video files
for i, start in enumerate(start_times):
    if start < video_duration:
        end = min(start + 120, video_duration)  # Split into 2-minute segments
        segment = original_clip.subclip(start, end)
        segment_filename = f"segment_{i + 1}.mp4"
        segment.write_videofile(segment_filename, codec="libx264")
        clips.append(segment_filename)

# Provide download links for each video segment
for i, clip_filename in enumerate(clips):
    print(f"Segment {i + 1}:")
    print(f"[Download Link](/{clip_filename})")

```
